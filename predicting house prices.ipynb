```python
import turicreate as tc
```

```python
sf=tc.SFrame("house_data")
```

```python
sf
```

```python
tc.show(sf["sqft_living"],sf["price"])
```

```python
tc.show(sf["sqft_living"][1:],sf["price"][1:])
```

```python
train_set,test_set=sf.random_split(.8,seed=0)
```

```python
print(train_set)
```

```python
sqft_model=tc.linear_regression.create(train_set,target='price',features=['sqft_living'])
```

```python
print(test_set['price'].mean())
```

```python
print(sqft_model.evaluate(test_set))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(test_set['sqft_living'],test_set['price'],'.',test_set['sqft_living'],sqft_model.predict(test_set),'-')
```

```python
sqft_model._get('coefficients')
```
